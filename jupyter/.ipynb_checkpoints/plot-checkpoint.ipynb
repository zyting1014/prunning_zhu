```python
y = ['1.16', '1.28', '1.24', '1.25', '1.22', '1.19', '1.31', '1.18', '1.14', '1.15', '1.13', '1.09', '1.04', '1.08', '1.04', '1.02', '1.03', '1.13', '0.94', '0.91']
```

```python
x =  ['1.00', '0.95', '0.90', '0.85', '0.81', '0.77', '0.73', '0.70', '0.66', '0.63', '0.59', '0.56', '0.53', '0.51', '0.48', '0.46', '0.43', '0.41', '0.39', '0.37']
```

```python
y = list(map(float, y))
x = list(map(float, x))
```

```python
plt.figure(figsize = (10,5))
plt.plot(x,y)
plt.xlim((max(x),min(x)))
plt.ylim((min(y) - 0.1,max(y) + 0.1))
plt.ylabel("推理时间")
plt.xlabel("压缩率")
plt.show()
```
